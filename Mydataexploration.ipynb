```python
##plotly for Total BsmtSF

import plotly.graph_objs as go
import plotly.plotly as py
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

xaxis=x['TotalBsmtSF']
yaxis=x['SalePrice']
trace1 = go.Scatter(
    y =yaxis,
    x=xaxis,
    mode='markers',
    marker=dict(
        size='16',
        color = np.random.randn(2000), #set color equal to a variable
        colorscale='Viridis',
        showscale=True
    )
)
data = [trace1]
layout= go.Layout(
    title= 'TotalBsmtSf vs SalesPrice',
    hovermode= 'closest',
    xaxis= dict(
        title= 'TotalBsmtSf',
        ticklen= 5,
        zeroline= False,
        gridwidth= 2,
    ),
    yaxis=dict(
        title= 'SalePrice',
        ticklen= 5,
        gridwidth= 2,
    ),
    showlegend= False
)
fig= go.Figure(data=data, layout=layout)
iplot(fig)

#fig= go.Figure(data=[trace1], layout=layout)
#iplot(data)
```
